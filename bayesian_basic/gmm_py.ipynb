```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import mixture, datasets
from sklearn.cluster import KMeans
plt.rcParams['figure.figsize'] = [16, 8]
plt.style.use("ggplot")
```

Create blobs data from datasets package

```python
X_blobs = datasets.make_blobs(n_samples=10000, random_state=2)
# random state = 3 (n-1 = 2)
```

```python
X = X_blobs[0] # extract values only
```

```python
X
```

```python
components = 3 
```

Show data

```python
plt.scatter(X[:, 0], X[:, 1], c = X_blobs[1] ,s=4, alpha = 0.8);
```

```python
X_blobs[1]
```

Fit Gaussian Mixture Model to data

```python
# Fit a GMM with estimation using 3 components
gmm = mixture.GaussianMixture(n_components=components).fit(X)
```

```python
prediction_gmm = gmm.predict(X)
probabilities_gmm = gmm.predict_proba(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=prediction_gmm, s=4, alpha=0.8);
```

```python
plt.scatter(X[:,0], X[:,1], c=probabilities_gmm, s=4, alpha=0.8);
```
